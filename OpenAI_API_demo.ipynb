```python
import os
from openai import OpenAI

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

client = OpenAI(
    # This is the default and can be omitted
    api_key=os.environ.get("OPENAI_API_KEY"),
)
```

```python
# llm_model = "gpt-3.5-turbo"
llm_model = "gpt-4o-mini" # smarter and cheaper
```

```python
def get_completion(prompt, model=llm_model):

    response = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model=model,
    # temperature=0,
)


    return response.choices[0].message.content
```

```python

prompt = """
Generate sarcastic statements and provide straightforward, literal translations for each one. 
Here are a few examples to guide you, Try to give different topics. 
Try not to use modal particle like 'Wow', 'Oh', 'Yay', 'Fantastic!'.
Make your writing style more various, especially for Sarcastic. 
:

---
Sarcastic: "It's been a blissful 2.5 hours of sleep tonight."
Translation: "I only got 2.5 hours of sleep last night and it wasn't enough"

Sarcastic: "90 percent of adulthood is just refilling your Brita pitcher."
Translation: "I have to refill my Brita pitcher too often."

Sarcastic: "I just absolutely LOVE how I've got to work outside for the next 3 days in the heatwave."
Translation: "I hate working outside in the heat."
---


Generate 300 more sarcastic statements along with their straightforward translations. Only return the results. The style of results should follow:

"
Sarcastic: "xxx"
Translation: "xxx"

Sarcastic: "xxx"
Translation: "xxx"

...

Sarcastic: "xxx"
Translation: "xxx"
"

"""
```

```python
response = get_completion(prompt)
```

```python
response
```

```python
def clean_text(text):
    text = text.replace('‘', "'").replace('’', "'")  # Replacing curly single quotes with straight ones
    text = text.replace('“', '"').replace('”', '"')  # Replacing curly double quotes with straight ones
    return text

response = clean_text(response)
```

```python
lines = response.strip().split("\n\n")[1:-1]
sarcastic_translation_pairs = []

for line in lines:
    parts = line.split("\n")
    if len(parts) < 2:
        print(parts)
    sarcastic = parts[0].replace('Sarcastic: ', '').strip('"')
    translation = parts[1].replace('Translation: ', '').strip('"')
    sarcastic_translation_pairs.append((sarcastic, translation))
```

```python
import csv

with open('sarcastic_translations.tsv', mode='a', newline='') as file:
    writer = csv.writer(file, delimiter='\t')  # Use tab as delimiter
    
    # Write headers
    # writer.writerow(["Sarcastic", "Translation"])

    # Write sarcastic/translation pairs
    writer.writerows(sarcastic_translation_pairs)
```
